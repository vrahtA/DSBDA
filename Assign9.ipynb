```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
print(titanic.head())
```

```python
# Histogram plot
age = titanic['Age'].ffill()
sns.histplot(age,bins=30,kde=True)
plt.title('Price Distribution')
plt.show()
```

```python
# Joint plot
sns.jointplot(x='Age',y='Fare',data=titanic,kind="scatter")
plt.show()
```

```python
# Pair plot for numerical columns
sns.pairplot(titanic[['Age', 'Fare', 'SibSp', 'Parch']])
plt.show()
```

```python
# Run plot (line plot) of age over time
sns.lineplot(x='Age', y='Fare', data=titanic)
plt.title('Line Plot of Age vs Fare')
plt.show()
```

```python
# Bar plot for average age by passenger class
sns.barplot(x='Pclass', y='Age', data=titanic)
plt.title('Average Age by Passenger Class')
plt.show()
```

```python
# Count plot for survival count by class
sns.countplot(x='Sex',hue='Survived', data=titanic)
plt.title('Survival Count by Passenger Class')
plt.show()
```

```python
# Strip plot for Age by Embarked
sns.stripplot(x='Embarked', y='Fare', data=titanic, jitter=True)
plt.title('Strip Plot of Age by Embarked')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(15,8))
sns.boxplot(data=titanic,x='Sex',y='Age',hue='Survived')
# Adding titles and labels
plt.title('Distribution of Age by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
# Facet grid for survival by class and sex
sns.FacetGrid(titanic, col='Survived', row='Pclass').map(sns.histplot, 'Sex')
plt.show()
```
